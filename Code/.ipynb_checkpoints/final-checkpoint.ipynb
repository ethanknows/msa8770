### construct the good and bad text data

#### function for get the text data from the dataframe

```python
def get_text(text):
    texts = []
    for i in top_ten_product:
        texts.append(text[text.ProductID == i]["ReviewContent"].str.cat(sep=' '))
    return texts
```

```python
good_text = get_text(good_raw)
good_text
```

```python
bad_text = get_text(bad_raw)
bad_text
```

# Data cleaning

### Lower the text and strip the punctuation

```python
def clean_text(text):
    lower = [doc.lower() for doc in text]
    cleaned_text = [doc.translate(None, string.punctuation) for doc in lower]
    return cleaned_text
```

```python
cleaned_text = clean_text(good_text)
```

### Part of Speech Tag

```python
def tag_text(text):
    tagged = [pos_tag(doc.split()) for doc in text] #use NLTK's part of speech tagger
    tag_list = ["JJ","JJR","JJS","NN","NNP","NNS","RB","RBR", \
                "RBS","UH","VB","VBD","VBG","VBN","VBP","VBZ"]
    new_tagged = [[word for word,pos in doc if pos in tag_list] for doc in tagged]
    tagged_text = [" ".join(doc) for doc in new_tagged]
    return tagged_text
```

```python
tagged_text = tag_text(cleaned_text)
```

### Tokenizing and Lemmatizing

```python
def token_lemma(text):
    tokens = [nltk.word_tokenize(doc) for doc in text]
    lemmatizer = WordNetLemmatizer()
    docs = [[lemmatizer.lemmatize(token.decode("utf-8")) for token in doc] for doc in tokens]
    return docs
```

```python
docs = token_lemma(tagged_text)
```

### Compute Bigrams

```python
def add_bigram(text):
    bigram = Phrases(text)
    for idx in range(len(text)):
        for token in bigram[text[idx]]:
            if '_' in token:
                text[idx].append(token)
    return text
```

```python
bi_docs = add_bigram(docs)
```

```python
i = 0
for word in bi_docs[0]:
    if "_" in word:
        i+=1
        print word,i
```

### Stop Words Removal 

```python
stop_list = stopwords.words("english")
```

```python
f_docs = [[w for w in doc if not w in stop_list] for doc in bi_docs]
```

# Model Training

```python
dictionary = corpora.Dictionary(f_docs)
# Filter out words that occur less than 1 documents, or more than 90% of the documents.
dictionary.filter_extremes(no_below=1, no_above=0.9)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in f_docs]
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

```python
# Train LDA model.
random.seed(1000)
# Set training parameters.
num_topics = 50
chunksize = 200
passes = 20
iterations = 4000

%time model = LdaModel(corpus=corpus, id2word=dictionary, \
                       chunksize=chunksize, \
                       alpha='auto', eta='auto', \
                       iterations=iterations, \
                       num_topics=num_topics, \
                       passes=passes)
```

```python
model.print_topics(num_topics=50, num_words=10)
```

```python
for i in range(10):
    print model.get_document_topics(corpus[i])
```
