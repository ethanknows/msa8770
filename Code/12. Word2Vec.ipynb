```python
from gensim.models import Word2Vec
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
# http://www.nltk.org/book/ch02.html
from nltk.corpus import brown
sentences = brown.sents()
```

```python
#Word2vec accepts several parameters that affect both training speed and quality.

#One of them is for pruning the internal dictionary. 
#Words that appear only once or twice in a billion-word corpus are probably uninteresting typos and garbage. 
#In addition, there’s not enough data to make any meaningful training on those words, 
#so it’s best to ignore them:

#model = Word2Vec(sentences, min_count=10)  # default value is 5
#A reasonable value for min_count is between 0-100, depending on the size of your dataset.

#Another parameter is the size of the NN layers, which correspond to the “degrees” 
#of freedom the training algorithm has:

#model = Word2Vec(sentences, size=200)  # default value is 100
#Bigger size values require more training data, but can lead to better 
#(more accurate) models. Reasonable values are in the tens to hundreds.

#The last of the major parameters (full list here) is for training parallelization, to speed up training:

#model = Word2Vec(sentences, workers=4) # default = 1 worker = no parallelization
#The workers parameter has only effect if you have Cython installed. 
#Without Cython, you’ll only be able to use one core because of the GIL 
#(and word2vec training will be miserably slow).

model = Word2Vec(sentences, min_count=100)
```

```python
model.save('brown_model')
print "Brown corpus model saved."
```

```python
model = Word2Vec.load("brown_model")
#words most similar to mother
print model.most_similar("mother")
print model.most_similar("great")
print model.most_similar("money")


#vector representation of word human
print len(model["human"])
print model.most_similar(positive=['woman','queen'], negative=['man'],topn = 1)
print model.similarity('woman', 'man')
```
